```python
# importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd

# set parameters
users = pd.read_csv ('accounts_twitter.csv', usecols= ['Arroba'])
keyword_n_hashtags_list = "#paulofreirelixo", "#paulofreireamavaocomunismo", "#paulofreireenergúmeno", "#paulofreireéumafraude", "#paulofreirepatronodofracasso", "#paulofreire100anosdedestruiçaodobrasil",  "#paulofreirecomunista", "#paulofreireeumidiota", "#paulofreiremediocre", "#paulofreirenaomerepresenta", "#paulofreire", "Paulo Freire"
start_date = '2012-01-01'
end_date = '2021-30-12'


# Creating list to append tweet data 
tweets_list1 = []
twitter_query = ""

for user in users.iterrows():
    for keyword in keyword_n_hashtags_list:
        twitter_query = keyword + " from:" + user[1]['Arroba'] + " since:" + start_date  
        print(twitter_query)

        # Using TwitterSearchScraper to scrape data and append tweets to list
        for i,tweet in enumerate(sntwitter.TwitterSearchScraper(twitter_query).get_items()): 
            tweets_list1.append([tweet.user.username, tweet.user.displayname, tweet.user.description, tweet.date.strftime("%d/%m/%Y"), tweet.content, tweet.hashtags, tweet.outlinks, tweet.media, tweet.replyCount, tweet.retweetCount, tweet.likeCount, tweet.id]) 


# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Username', 'Display Name', 'Account description', 'Datetime', 'Text', 'Hashtags', 'Outlinks', 'Medias', 'Replies', 'Retweets', 'Likes', 'Tweet Id'])

# Delete duplicate tweets
tweets_df1.drop_duplicates(subset='Tweet Id', keep = False, inplace = True)

# Save data frame in an excel.
tweets_df1.to_excel('tweets.xlsx', index = False)      
```
